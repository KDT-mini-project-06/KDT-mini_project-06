```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Data
data = {
    'years': ['1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003',
              '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
              '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023'],
    'values': [12.766667, 6.700000, 7.400000, 5.100000, 1.100000, 0.966667, 3.633333,
               4.600000, 2.133333, 0.633333, 5.466667, 3.433333, 5.533333, 3.933333,
               5.933333, 0.766667, 6.800000, 2.866667, 5.700000, 8.500000, 2.966667,
               4.033333, 7.700000, 5.033333, 10.033333, 4.900000, 3.100000, 3.933333,
               4.433333, 4.700000]
}
df = pd.DataFrame(data)

# Linear regression
X = df.index.astype(int).values.reshape(-1, 1)
y = df['values'].values
reg = LinearRegression().fit(X, y)

# Predict values
y_pred = reg.predict(X)

# Plot
plt.figure(figsize=(12, 6))
plt.scatter(df['years'], df['values'], color='skyblue', label='폭염지수')
plt.plot(df['years'], y_pred, color='r', linestyle='--', label='회귀 직선')
plt.title('폭염')
plt.xlabel('years')
plt.ylabel('폭염지수')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 주어진 데이터
data = {
    '연도/월': ['1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003',
               '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
               '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021', '2022'],
    
    '합계': [5, 3, 2, 4, 2, 5, 5, 1, 4, 4, 5, 1, 3, 3, 1, 0, 3, 3, 5, 3, 4, 4, 2, 3, 5, 7, 4, 3, 5]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# 연도/월 열을 인덱스로 설정
df.set_index('연도/월', inplace=True)

# X와 y 값 설정
X = np.arange(len(df)).reshape(-1, 1)
y = df['합계']

# 선형 회귀 모델 학습
model = LinearRegression().fit(X, y)

# 회귀 직선 예측
y_pred = model.predict(X)

# 그래프 그리기
plt.figure(figsize=(12, 6))

# 막대 그래프 그리기
df.plot(kind='bar', stacked=True, ax=plt.gca(), color='skyblue', label='태풍 발생 횟수')
plt.ylabel('발생 횟수')

# 회귀 직선 그리기
plt.plot(df.index, y_pred, color='r', linestyle='--', label='회귀 직선')

# 사과 생산량 그래프 추가
#df_apple['대구경북'].plot(ax=plt.gca(), label='사과 생산량', color='green', marker='o')
#plt.ylabel('단위:톤(t)')

plt.title('연도별 태풍 발생 횟수')
plt.xlabel('연도')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
