```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import koreanize_matplotlib

%precision 3
%matplotlib inline
```

```python
# 연도별 평균 강수량 회귀직선, 월별 평균 강수량 or 연도별 평균 강수량, 
rain_df = pd.read_excel('../DATA/경북 강수량 데이터 19940101-20231231.xlsx')
rain_df
```

```python
from sklearn.linear_model import LinearRegression

# 임의의 연도별 강수량 데이터 생성
years = np.array([1994, 1995, 1996, 1997, 1998, 1999, 2000,
                  2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
                  2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023])
mean_rainfall = np.array([706.9, 812.3, 947.9, 1135.4, 1603.7, 1454.6, 1103.1,
                    875.5, 1359.9, 1847.9, 1327.2, 1097.5, 1361.8, 1282.5, 850.1, 897.2,
                    1056.1, 1307, 1244.9, 992.4, 1152.9, 801, 1209.5, 835.3, 1330.5, 1128.4, 1333.2, 1157.5, 856.8, 1557.3])

# 데이터를 2차원 배열로 변환
years = years.reshape(-1, 1)

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(years, mean_rainfall)

# 회귀 계수 및 절편 확인
slope = model.coef_[0]
intercept = model.intercept_
print("회귀 계수:", slope)
print("절편:", intercept)

# 회귀직선 시각화
plt.figure(figsize=(10, 6))
plt.scatter(years, mean_rainfall, color='blue', label='연도별 강수량')
plt.plot(years, model.predict(years), color='red')
plt.title("경북 연도별 누적 강수량")
plt.xlabel("연도")
plt.ylabel("강수량(mm)")
plt.legend()
plt.grid(True)
plt.show()
```
